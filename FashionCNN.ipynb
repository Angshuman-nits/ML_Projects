# ***Building CNN***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

***Change the dimension of the data(rn we've 3D data)***

```python
x_train.ndim
```

```python
x_train.shape
```

***we need to convert it to 4D***

```python
x_train=np.expand_dims(x_train,-1)
x_test=np.expand_dims(x_test,-1)
```

```python
x_train.shape
```

```python
x_train.ndim
```

***Feature Scaling (so that every pixel values lie between 0 to 1)***

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
x_train[0]
```

***Split dataset for cross validation(trainning ke phase mei testing)***

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2,random_state=2020)
```

```python
x_train.shape, y_train.shape
```

```python
x_val.shape, y_val.shape
```

# ***CNN MODEL BUILDING***

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten,Dropout
```

```python
model=Sequential([
    Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)),
    MaxPool2D(pool_size=(2,2)),  #to downsample  our images
    Flatten(),  #to flatten images into a single vector(or layer)
    Dense(128,activation='relu'),               #(input layer)
    Dense(10,activation='softmax')  #softmax points to probability (output layer)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10,validation_data=(x_val,y_val))
```

***TO TEST THE MODEL***

```python
x_test.shape
```

```python
model.predict(np.expand_dims(x_test[0],axis=0)).round(2)        #we have to give data in 4D
#round(2) for rounding up to 2 decimal places
```

```python
np.argmax(model.predict(np.expand_dims(x_test[0],axis=0)).round(2))
```

```python
y_test[0]      #to cross check the output 9
```

```python
y_pred=model.predict(x_test).round(2)   #x_test is 4D already
y_pred
```

```python
model.evaluate(x_test,y_test)
```

***to visualize the output***

```python
plt.figure(figsize=(16,15))
j=1
for i in np.random.randint(0, 1000,60):
  plt.subplot(10,6,j); j+=1
  plt.imshow(x_test[i].reshape(28,28), cmap = 'gray')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[y_test[i]], y_test[i], class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
  plt.axis('off')
  plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
```

```python
plt.figure(figsize=(16,9))
y_pred_labels = [np.argmax(label) for label in y_pred]
cm = confusion_matrix(y_test, y_pred_labels)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues' , xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
cr=classification_report(y_test,y_pred_labels)  #only when we have classification type problem
print(cr)
```

# ***CNN building complex models***

image.png

```python
from tensorflow.keras.layers import MaxPooling2D
```

```python
cnn_model2 = Sequential([
                         Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         MaxPooling2D(pool_size=(2,2)),
                         Conv2D(filters=64, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         MaxPooling2D(pool_size=(2,2)),
                         Flatten(),
                         Dense(units=128, activation='relu'),
                         Dropout(0.25),
                         Dense(units=256, activation='relu'),
                         Dropout(0.25),
                         Dense(units=128, activation='relu'),
                         Dense(units=10, activation='softmax')
                         ])
```

```python
cnn_model2.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn_model2.fit(x_train, y_train, epochs=20, batch_size=512, verbose=1, validation_data=(x_val, y_val))
```

```python
cnn_model2.evaluate(x_test, y_test)
```

# ***Very complex models***

```python
from tensorflow.keras.layers import Dropout
```

```python
cnn_model3 = Sequential([
                         Conv2D(filters=64, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         MaxPooling2D(pool_size=(2,2)),
                         Conv2D(filters=128, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         MaxPooling2D(pool_size=(2,2)),
                         Conv2D(filters=64, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         MaxPooling2D(pool_size=(2,2)),
                         Flatten(),
                         Dense(units=128, activation='relu'),
                         Dropout(0.25),
                         Dense(units=256, activation='relu'),
                         Dropout(0.5),
                         Dense(units=256, activation='relu'),
                         Dropout(0.25),
                         Dense(units=128, activation='relu'),
                         Dropout(0.10),
                         Dense(units=10, activation='softmax')
                         ])
```

```python
cnn_model3.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Train the Model
cnn_model3.fit(x_train, y_train, epochs=50, batch_size=512, verbose=1, validation_data=(x_val, y_val))
```

```python
cnn_model3.evaluate(x_test, y_test)
```
